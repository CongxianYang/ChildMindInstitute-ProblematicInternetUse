```python
# Read in the Test CSV Dataset
df = pd.read_csv('Data\child-mind-institute-problematic-internet-use\\test.csv')
df_test = pd.DataFrame(df)
```

```python
import copy
for item in df_train.columns:
    if df_train[item].dtype=='object':
        df_train[item]=df_train[item].fillna('missing')
    else:
        df_train[item]=df_train[item].fillna(df_train[item].mean())
df_train=copy.deepcopy(df_train)
df_train
```

```python
for item in df_test.columns:
    if df_test[item].dtype=='object':
        df_test[item]=df_test[item].fillna('missing')
    else:
        df_test[item]=df_test[item].fillna(df_test[item].mean())
df_test=copy.deepcopy(df_test)
df_test
```

```python
columns_with_missing = df_train.columns[df_train.isnull().any()]
columns_with_missing
```

```python
X = df_train.drop(['id', 'sii'], axis=1)
X = pd.get_dummies(X)
Y=df_train['sii']
X
```

```python
Y
```

```python
##测试集
X_test=df_test.drop(['id'], axis=1)
X_test=pd.get_dummies(X_test)
X_test
```

```python
common_columns =X.columns.intersection(X_test.columns)
X = X[common_columns]
X_test = X_test[common_columns]
# 检查对齐后的字段数量
print(f"Number of columns in aligned training set: {len(X.columns)}")
print(f"Number of columns in aligned test set: {len(X.columns)}")
# 检查对齐后的字段名称
print(f"Columns in aligned training set: {X.columns.tolist()}")
print(f"Columns in aligned test set: {X.columns.tolist()}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import pandas as pd
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score
import pickle
from tqdm import tqdm
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1234)
```

```python
pipelines = {
    'rf': make_pipeline(StandardScaler(), RandomForestRegressor(random_state=1234)),
    'gb': make_pipeline(StandardScaler(), GradientBoostingRegressor(random_state=1234))
}
# 定义网格搜索的参数
grid = {
    'rf': {
        'randomforestregressor__n_estimators': [100, 200, 300, 400],
        'randomforestregressor__max_depth': [None, 10, 20, 30],
        'randomforestregressor__min_samples_split': [2, 5, 10],
        'randomforestregressor__min_samples_leaf': [1, 2, 4]
    },
    'gb': {
        'gradientboostingregressor__n_estimators': [100, 200, 300, 400],
        'gradientboostingregressor__learning_rate': [0.01, 0.1, 0.2],
        'gradientboostingregressor__max_depth': [3, 5, 10],
        'gradientboostingregressor__min_samples_split': [2, 5, 10],
        'gradientboostingregressor__min_samples_leaf': [1, 2, 4]
    }
}
```

```python
# 进行网格搜索
fit_models = {}
for name, pipe in pipelines.items():
    grid_search = GridSearchCV(pipe, grid[name], cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
    grid_search.fit(x_train, y_train)
    print(f"Best parameters for {name}: {grid_search.best_params_}")
    print(f"Best score for {name}: {-grid_search.best_score_}")
    fit_models[name] = grid_search
```

```python
# Evaluate the performance of the model
# 评估模型
for name, model in fit_models.items():
    # 使用最佳模型进行预测
    y_pred = model.predict(x_test)
    # 计算评估指标
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    # 打印评估指标
    print(f"\nEvaluation for {name}:")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"R² Score: {r2:.4f}")
with open('models\\gradientboostingregressor.pkl', 'wb') as f:
    pickle.dump(fit_models['gb'], f)
# with open('models\\randomforestregressor.pkl', 'wb') as f:
#     pickle.dump(fit_models['rf'], f)
```

```python
yhat_test = fit_models['gb'].predict(X_test)
submission = pd.DataFrame([X_test['id'], yhat_test]).T
submission.columns = ['id', 'iss']
submission.to_csv('submission\kaggle_submission.csv', index=False)
```
